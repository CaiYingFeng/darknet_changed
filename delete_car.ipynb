```python
from PIL import Image
from PIL import ImageDraw
from PIL import ImageFont
import os
def delete_car(img,x_min,x_max,y_min,y_max):
#     path_dir='/home/autolab/caiyingfeng/darknet/data/out/'
#     filename='00002550.jpg'
    # 拼接图片目录
#    path=os.path.join(path_dir,filename)
    # 打开图片
#    img = Image.open(path)
    # 图片宽度
#    width = img.size[0]
    # 图片高度
#    height = img.size[1]

#    print(width)
#    print(height)
    RGB_data=(0,0,0)
    for x in range(x_min,x_max):
        for y in range(y_min,y_max):
            # 将该区域颜色改成背景色
            img.putpixel((x,y),RGB_data)          
    img = img.convert("RGB")#把图片强制转成RGB
    
    return img

 # 另存图片
# new_path=os.path.join(path_dir,"new"+filename)
# img.save(new_path)#保存图片

def change_img(img,x_min,x_max,y_min,y_max):
    img=img.convert('1')
    
    for x in range(x_min,x_max):
        for y in range(y_min,y_max):
            # 将该区域颜色改成背景色
            img.putpixel((x,y),0)          
    #img = img.convert("RGB")#把图片强制转成RGB
    
    return img
```
